```python
import pandas as pd

df = pd.read_csv('customer_shopping_behavior.csv')


df.head()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
from sqlalchemy import Table, Column, Integer, Float, String, MetaData
```

```python
metadata = MetaData()

customer_table = Table(
    "customer_datatype",
    metadata,

    Column("customer_id", Integer),
    Column("age", Integer),
    Column("gender", String(20)),
    Column("item_purchased", String(100)),
    Column("category", String(50)),
    Column("purchase_amount", Integer),
    Column("location", String(100)),
    Column("size", String(20)),
    Column("color", String(30)),
    Column("season", String(30)),
    Column("review_rating", Float),
    Column("subscription_status", String(20)),
    Column("shipping_type", String(50)),
    Column("discount_applied", String(10)),
    Column("previous_purchases", Integer),
    Column("payment_method", String(50)),
    Column("frequency_of_purchases", String(50)),
    Column("age_group", String(50)),
    Column("purchase_frequency_days", Integer)
)
```

```python
metadata.create_all(engine)
print("Table created successfully in Oracle!")
```

```python
df.to_sql(
    "customer_datatype",
    engine,
    if_exists="append",
    index=False
)
```
